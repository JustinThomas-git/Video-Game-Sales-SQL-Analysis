```python
import requests
from bs4 import BeautifulSoup
import pandas as pd
from sqlalchemy import create_engine
```

```python
url = 'https://www.vgchartz.com/games/games.php?console=PS4'
```

```python
response = requests.get(url)
```

```python
response
```

```python
response.text
```

```python
soup = BeautifulSoup(response.text, 'html.parser')
```

```python
print(soup.prettify())
```

## Games were under same tag/attribute, but different value. Found both to get all page results ##

```python
scores = soup.findAll('tr', attrs={'style': 'background-image:url(../imgs/chartBar_large.gif); height:70px'})
```

```python
scores2 = soup.findAll('tr', attrs={'style': 'background-image:url(../imgs/chartBar_alt_large.gif); height:70px'})
```

```python
print(scores2)
```

```python
print(scores)
```

```python
score_details = {
    'Name':[],
    'Publisher':[],
    'CriticScore':[]
}

for score in scores:
    
    name = score.find('a', attrs={'style':'color:#008fd4'}).text
    score_details['Name'].append(name)
    print('Name:', name)
    
    publisher = score.find('td', attrs={'width':'100'}).text
    score_details['Publisher'].append(publisher)
    print('Publisher:', publisher)
    
    cscore = score.findAll('td', attrs={'align':'center'})
    score_details['CriticScore'].append(cscore[2].text)
    print('CriticScore:', cscore[2].text)
    
    
    print('-'*75)
    
df = pd.DataFrame(score_details)
df.to_sql('GameScore', con=engine, if_exists='append', index=False)
```
