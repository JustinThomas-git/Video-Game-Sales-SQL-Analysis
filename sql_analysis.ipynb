### 2. Find games with more than 1.5million sold in NA ###

```python
%%sql
SELECT 
    DISTINCT(Game), 
    NASales,
    Releaseyear
FROM GameSales 
WHERE NASales > 1.5
ORDER BY NASales DESC;
```

### 3. Find duplicate values ###

```python
%%sql
SELECT 
    Game, 
    COUNT(*)
FROM GameSales 
GROUP BY Game 
HAVING COUNT(*) > 1;
```

There are duplicate game titles, use DISTINCT

### 4. List the Publishers and number of games for each ###

```python
%%sql 
SELECT 
    Publisher,
    COUNT(DISTINCT(Game)) AS NumOfGames
FROM GameSales
GROUP BY Publisher
ORDER BY COUNT(DISTINCT(Game)) DESC;
```

Lists all game publishers and amount of games under each publisher.

### 5. Identify the range of release years in the data ###

```python
%%sql
SELECT 
    MAX(ReleaseYear),
    MIN(ReleaseYear)
FROM GameSales;
```

# Games with the highest sales in North America > than the average #

### Primary Question: Which PS4 video games and their genres are the highest sellers in North America? ###

Getting the above average sellers and genre to see what titles have the highest sales.

CTE, Subquery, JOIN

```python
%%sql
WITH GameSalesCTE AS (
    SELECT 
        DISTINCT(Game), 
        UCASE(GenreName),
        NASales
    FROM GameSales s
    JOIN Genre g
        ON s.GenreID = g.GenreID
    ORDER BY NASales DESC
)
SELECT *,
    (
        SELECT AVG(NASales)
        FROM GameSalesCTE
    ) AS AvgNASales
FROM GameSalesCTE
WHERE NASales > (
    SELECT AVG(NASales)
    FROM GameSalesCTE
);
```

Shooter and Action games dominate the North American Sales, with Sports being the 3rd most popular amongst consumers.
Only three games sold more than 5 million copies. These genre's are most popular and should be prioritized.

TOP 3: Black Ops 3, GTAV, RDR2

# Percentage of Global Sales per Genre #

Count the games in each genre and their total sales compared to the Global Sales. 

GROUP BY, JOIN

```python
%%sql
SELECT 
    COUNT(DISTINCT(Game)) AS GamesInGenre, 
    UCASE(g.GenreName) AS Genre,
    SUM(NASales) AS TotalNASales,
    (SUM(NASales) / SUM(GlobalSales) * 100) AS PercentOfGlobalSales
FROM GameSales s
JOIN Genre g
    ON s.GenreID = g.GenreID
GROUP BY GenreName
ORDER BY GamesInGenre DESC;
```

Action games has more than double the amount of titles as Shooter games. Sports in 3rd with 5 less titles than Shooter. 45% of Fighting games are sold in North America. For the Fighting genre, I would increase efforts of advertising these games to other regions like Japan and the EU.

# Highest Scoring Games # 

Get the games and critic scores, ranked from High - Low

CASE

```python
%%sql
SELECT 
    DISTINCT(Name),
    CriticScore,
    CASE 
        WHEN CriticScore < 8 THEN 'Low'
        WHEN CriticScore BETWEEN 8 AND 9.2 THEN 'Average'
        WHEN CriticScore BETWEEN 9.3 AND 9.9 THEN 'High'
        ELSE 'N/A'
    END AS ScoreLevel
FROM GameScore
WHERE CriticScore != 0
ORDER BY CriticScore DESC;
```

2 0f 3 top sellers had a high critic score of 9.7 and above(RDR2, GTAV)
There are no Shooter games with a High score.


The #1 seller COD:Black Ops 3 has an average score of 8.1.
COD: Ghosts was the lowest selling game of the Call of Duty franchise and also had the lowest critic score.

Although critic scores do not always line up with the consumers choice, they can be indictive of a good or bad game
